### Basic Data cleaning

```python
text = "     <p>Hello!!! 😊 Visit https://example.com</p> NLP is FUN!!! #AI @2026 "
```

```python
def clean_text (text):
  text = text.lower()
  #remove html tags
  text = re.sub(r'<.*?>'," ",text)
  #remove urls
  text = re.sub(r'http\S+|www\S+',' ',text)
  #remove emojies
  text = re.sub(r'[^x00-\x7F]+',' ',text)
  #remove special characters
  text = re.sub(r'[^a-z0-9\s]',' ',text)
  #remove white spaces
  text = re.sub(r'\s+',' ',text).strip()
  return text
```

```python
clean_text(text)
```

# Stopwords

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
text = 'NLP is easy and very interesting to explore more about it'
words = word_tokenize(text)
[x for x in words if x not in stop_words]
```

# Stemming

```python
from nltk.stem import PorterStemmer
text = "playing played plays studies studying"
words = word_tokenize(text)
ps = PorterStemmer()
[ps.stem(x) for x in words]
```

# **POS Tagging**

```python
from nltk import pos_tag
text = ' NLP is easy and very intereesting to explore more about it'
words = word_tokenize(text)
pos_tag(words)
```

# Name Entity Recognition (NER)

```python
from nltk import ne_chunk
text = "John is working at Google company as HR in Bangalore"
words = word_tokenize(text)
tag = pos_tag(words)
ner = ne_chunk(tag)
print(ner)
```

```python
import spacy
model = spacy.load('en_core_web_sm')
```

### **Word tokenization**

```python
text = 'NLP is easy'
words = model(text)
[x for x in words]
```

### **Sentence Tokenization**

```python
text = 'NLP is easy. Python is hard'
sentences = model(text)
[x.text for x in sentences.sents]
```

### **Stop words**

```python
text = 'NLP is very easy and interesting'
words = model(text)
[x.text for x in words if not x.is_stop]
```

### **Lemmatization**

```python
text = 'playing played plays studies studying better good'
words = model(text)
[x.lemma_ for x in words]
```

### **POS Tagging**

```python
text = ' I have started studying NLP and its very interesting'
words = model(text)
{x:x.pos_ for x in words if not x.is_stop}
```

### **NER**

```python
text = "John is working at Google company as HR in Bengaluru"
words = model(text)
{x:x.label_ for x in words.ents}
```
